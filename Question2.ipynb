```julia
using Plots

γ_c(k, r, ae, ai) = (r^2 + (1 + r^2) * k^2 + k^4) / ((ae - ai) * r^2 + (ae - ai * r^2) * k^2)

r_range = [1/4,1/2, 3/4, 1]
k = 0:0.01:2.0
ae = 6
ai = 5

plot(k, [γ_c.(k, r, ae, ai) for r in r_range], ylim = (0,1.5), label = ["r = 1" "r = 1/4" "r = 3/4" "r = 1/2"], xlabel = 'k', ylabel = 'γ', lw = 2,
legend = :topleft)
```
